# Escalonamento Multidimensional

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.manifold import MDS
from sklearn.metrics import pairwise_distances
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.preprocessing import StandardScaler


ex1_df = pd.read_excel('Escalonamento exercicio1.xlsx')
ex2_df = pd.read_excel('Escalonamento exercicio2.xlsx')
ex3_df = pd.read_excel('Escalonamento exercicio3.xlsx')
ex4_df = pd.read_excel('Escalonamento exercicio4.xlsx', index_col=0)
ex5_df = pd.read_excel('Escalonamento exercicio5.xlsx')
ex6_df = pd.read_excel('Escalonamento exercicio6.xlsx')
paises_df = pd.read_excel('paises.xlsx')


class MDSAnalysis:
    def __init__(self, df, n_components=2, dissimilarity='precomputed', distance_metric='euclidean', distance_method='precomputed'):
        self.df = df
        self.n_components = n_components
        self.dissimilarity = dissimilarity
        self.distance_metric = distance_metric
        self.distance_method = distance_method
        self.stress, self.low_dim_data = self.compute_stress()

    def compute_stress(self):
        dist_matrix = self.df.values
        mds = MDS(n_components=self.n_components, dissimilarity=self.dissimilarity, random_state=0, normalized_stress="auto")
        low_dim_data = mds.fit_transform(dist_matrix)

        orig_distances = euclidean_distances(dist_matrix).flatten() if self.dissimilarity == 'euclidean' else dist_matrix.flatten()
        # orig_distances = self.calculate_distance()
        low_dim_distances = pairwise_distances(low_dim_data, metric=self.distance_metric).flatten()
        stress = np.sqrt(np.sum((orig_distances - low_dim_distances)**2) / np.sum(orig_distances**2))

        return stress, low_dim_data

    def calculate_distance(self):
        if self.distance_method == 'hamming':
            return pairwise_distances(self.df.values, metric=self.distance_metric)
        elif self.distance_method == 'euclidean':
            return euclidean_distances(self.df.values)
        elif self.distance_method == 'precomputed':
            return self.df.values.flatten()
        else:
            raise ValueError('Dissimilarity must be "euclidean" or "precomputed"')

    def plot_low_dim_data(self):
        plt.figure(figsize=(10, 10))
        plt.scatter(self.low_dim_data[:, 0], self.low_dim_data[:, 1], c='r', s=100)
        values_to_iterate = self.df.index if self.dissimilarity == 'euclidean' else self.df.columns
        for i, txt in enumerate(values_to_iterate):
            plt.annotate(txt, (self.low_dim_data[i, 0], self.low_dim_data[i, 1]), fontsize=20)
        plt.show()

    def analyze_quality_of_fit(self):
        quality_ranges = {
            0.05: "Excelente",
            0.1: "Muito bom",
            0.2: "Bom",
            0.3: "Aceitável"
        }

        for limit in sorted(quality_ranges.keys()):
            if self.stress < limit:
                return f"Qualidade do ajuste: {quality_ranges[limit]} (Stress: {self.stress:.4f})"

        return f"Qualidade do ajuste: Ruim (Stress: {self.stress:.4f})"

    def analyze_stress(self):
        print(f'Stress: {self.stress*100:.3f}%')
        print(self.analyze_quality_of_fit())
        self.plot_low_dim_data()
```
